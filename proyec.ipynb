```python
import json
import pandas as pd
import gzip
import ast

# Reemplaza 'ruta_del_archivo.json.gz' con la ruta real de tu archivo JSON comprimido en formato gzip
archivo_json = 'C:\\Users\\mikej\\Downloads\\steam_games.json.gz'

files_games = []

with gzip.open(archivo_json, 'rt', encoding='utf-8') as f:
    for line in f:
        data = json.loads(line)
        files_games.append(data)

df_steam_games = pd.DataFrame(files_games)
df_steam_games
```

```python
#Elimino valores nulos:
umbral = 3
df_steamgames_sin = df_steam_games.dropna(thresh=len(df_steam_games.columns) - umbral + 1)
```

```python
reviews_path = 'C:\\Users\\mikej\\Downloads\\user_reviews.json.gz'
reviews = []
with gzip.open(reviews_path, 'r') as f:
    for line in f.readlines():
        reviews.append(ast.literal_eval(line.decode("utf8")))

        

df_reviews = pd.DataFrame(reviews)
df_reviews
```

```python
#Y para desanidarlo para este caso, por ejemplo algo asÃ­: 

reviews = pd.json_normalize(df_reviews['reviews'].apply(lambda x: x[0] if x else None))

df_reviews = pd.concat([df_reviews, reviews], axis=1)

df_reviews = df_reviews.drop('reviews', axis=1)
df_reviews
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Inicializo el analizador de sentimientos de VADER
sia = SentimentIntensityAnalyzer()

 #Creo la funcion para obtener el analisis de sentimiento:
def obtener_sentimiento(texto):
    if isinstance(texto, str):
        sentiment_scores = sia.polarity_scores(texto)
        compound_score = sentiment_scores['compound']
        
        if compound_score >= 0.05:
            return 2  # Positivo
        elif compound_score <= -0.05:
            return 0  # Malo
        else:
            return 1  # Neutral
    else:
        return 1  # Si no hay reseÃ±a, valor neutral

# Aplico la funciÃ³n de anÃ¡lisis de sentimiento a la columna 'review' y creo la columna 'sentiment_analysis'
df_reviewsok_sin['sentiment_analysis'] = df_reviewsok_sin['review'].apply(obtener_sentimiento)

# Elimino la columna 'review'
df_reviewsok_sin.drop(columns=['review'], inplace=True)
```
